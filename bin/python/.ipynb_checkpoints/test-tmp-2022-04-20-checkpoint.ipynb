la cellule ci-dessous définit une fonction purement Python pour calculer l'entropie

```python
from scipy.spatial import cKDTree
from scipy.special import psi

# pure Python function (to check against the C code)
# the convention for dimensions ordering is different from the cython code
def entropy_nns(data, k):
    """
    Function to compute the Shannon entropy using a nearest neighbour search algorithm
    data has Npts lines (in time) et m columns (its dimensionality)
    """
    
    if (data.shape[0]<data.shape[1]): data=data.transpose()
    Npts = data.shape[0]
    m    = data.shape[1]
   
    kd=cKDTree(data) # Construct the tree
    
    # look for the k+1 nearest neighbours
    # the +1 is necessary because the point itself is considered as his own neighbour
    # p=inf tells to use the infinite norm (max of difference between coordinates)
    dist=kd.query(data,k+1,p=numpy.inf)[0]
    h=numpy.log(dist.max(axis=1)).sum()/Npts # moyenne des logs des diametres des boules
    
    # print  "h ,nx :",log(dist.max(axis=1)).sum(),data.shape[0]
    h = h*m - psi(k) + psi(Npts) + m*numpy.log(2.) # additional terms
    
    return h
```

# Entropie de Shannon et de Renyi

```python
# testing the Shannon entropy (Cython vs pure Python codes)
#
# parameters
npoints = 1000000
ndim  = 1
sigma = 1.
k     = 5
m_embed = 1
stride  = 10
tau_Theiler = 200

# expected data is of size (npts,m), i.e., the time dimension is the first one
data = normal(loc = 1.0, scale=sigma, size=(ndim, npoints))

print("normal distribution, ", npoints, "points,", ndim, "dimensions,", " stride=", stride)
print("Theoretical entropy :        H=", 0.5*numpy.log((2.*pi*numpy.exp(1)*sigma**2.)**ndim))

# "Pure python with cKDTree : "
t1=time()
res=entropy_nns(data, k)
t1=time()-t1;
print("Pure Python (with cKDTree) : H=", res, "\t\tTime :",t1)

# "ANN with Cython : "
t1=time()
res=entropy.compute_entropy(data, m_embed, stride, k=k)
t1=time()-t1;
print("Cython code old            : H=", res, "\t\tTime :",t1)
entropy.get_last_info(1)
res=entropy.compute_entropy(data, m_embed, stride, stride, k=k)
print()

# "ANN with Cython : "
t1=time()
res=entropy.compute_entropy_increments(data, m_embed, stride, k=k)
t1=time()-t1;
print("Cython code old increments : H=", res, "\t\tTime :",t1)
entropy.get_last_info(1)
print()

# new code with random set of realizations:
t1=time()
res=entropy.compute_entropy(data, m_embed, stride, tau_Theiler=-1, N_eff=1000, N_real=stride, k=k)
t1=time()-t1;
print("Cython code new            : H=", res, "\t\tTime :",t1)
entropy.get_last_info(1)
print()

H_S=res
```

```python
# signal x must be of shape (n, npts) (new convention)
def FIR_filter(x, T_integration):
    ''' function to filter in time (low pass) by local averaging :
    '''
    stride_f = T_integration*0.8 # overlap
    stride = int(stride_f)
    
    x_f = x[0,0:-1-T_integration];
    for i in range(1,T_integration):
        x_f  = x_f + x[0,i:-1-T_integration+i];
    x_f = x_f / T_integration;
           
    return tools.reorder(x_f)
```
