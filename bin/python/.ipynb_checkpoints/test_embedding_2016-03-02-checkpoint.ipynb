```python
from scipy import *
from scipy.spatial import *
from scipy.special import psi


def entropy_nns(data,k):
	"""
	Fonction pour calculer l'entropie avec un algo de recherche des plus proches voisins.
	data a N lignes (correspondants aux N points) et K colonnes (correspondants aux K dimensions)
	"""
	
	kd=cKDTree(data) # Construct the tree
	
	# look for the k+1 nearest neighbours
	# the +1 is necessary because the point itself is considered as his own neighbour
	# p=inf tells to use the infinite norm (max of difference between coordinates)
	dist=kd.query(data,k+1,p=inf)[0]
	
	h=log(dist.max(axis=1)).sum()/data.shape[0] # moyenne des logs des diametres des boules
	
	# print  "h ,nx :",log(dist.max(axis=1)).sum(),data.shape[0]
	h = h*data.shape[1] - psi(k) + psi(data.shape[0]) + data.shape[1]*log(2.) # additional terms
	
	return h
```

```python
from time import time
from numpy.random import normal # http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html

# params
m_range = np.arange(4)+1
N_range = 2**(np.arange(7)+7) # nb of points in time
Nrl  = 10   # nb of realisations
sigma= 1.
k    = 7
res_python = np.zeros((Nrl, N_range.shape[0], m_range.shape[0]))
res_nns_py = np.zeros((Nrl, N_range.shape[0], m_range.shape[0]))
res_ann_sw = np.zeros((Nrl, N_range.shape[0], m_range.shape[0]))
res_nns_sw = np.zeros((Nrl, N_range.shape[0], m_range.shape[0]))
res_bins   = np.zeros((Nrl, N_range.shape[0], m_range.shape[0]))

H0 = 0.5*np.log((2*pi*exp(1)*sigma**2.))
#print "normal distribution,", npoints, "points,", ndim, "dimensions."
print "Theoretical entropy per dimension : H=", H0

i_N=0;
for Npts in N_range:
    for m in m_range:
        print "working on embedding",m,"with",Npts,"points in time"
        for i in np.arange(Nrl):
            data = normal(loc = 1.0, scale=sigma, size=(Npts, m));

# "Pure python with cKDTree : "
# t1=time()
            res_python[i, i_N, m-1]=entropy_nns(data, k)
# t1=time()-t1;
# print "Pure Python (with cKDTree) : H=", res, "\tTime :",t1

           res_nns_py[i, i_N, m-1]=ann.compute_entropy(data,1,1, k)

# "C wrapper with kd-tree if n>1 : "
# t1=time()
       #     ret, res_nns_sw[i, i_N, m-1] =entropy.compute_entropy_nns(data,1,1, k)
# t1=time()-t1
# print "C (compute_entropy_nns) :    H=", res, "\tTime :",t1

# "C wrapper with kd-tree if n>1 : "
# t1=time()
            ret, res_ann_sw[i, i_N, m-1]=entropy.compute_entropy_ann(data,1,1, k)
# t1=time()-t1
# print "C (compute_entropy_ann) :    H=", res, "\tTime :",t1

# "C wrapper with bins if n>1 : "
# t1=time()
       #     res_bins[i, i_N, m-1] = entropy.compute_entropy_bins(data, 1, 7)  # stride=1 (UNUSED VARIABLE !!!), 7 bins
# t1=time()-t1
# print "C (compute_entropy_bins) :   H=", res, "\tTime :",t1
    i_N += 1

#np.save("res_python.pyd", res_python)
#np.save("res_nns_py.pyd", res_nns_py)
np.save("res_ann_sw.pyd", res_ann_sw)
#np.save("res_nns_sw.pyd", res_nns_sw)
#np.save("res_bins.pyd",   res_bins  )
```

```python
def plot_res_vs_N(fig, N_range, res, my_title, do_legend):
    r_m = np.mean(res,0)
    r_s = np.std (res,0)
    
    for m in m_range:
        my_text= "m="+str(m)
        fig.plot([0,N_range[-1]], [H0,H0], "--k")

        h1  = fig.errorbar(N_range, r_m[:,m-1]/m, yerr=r_s[:,m-1]/m, 
                       marker='s', label=my_text)

#figa.set_xlim(0, numpy.max(g_n))  
#figa.set_ylim(0, 0.75)  
    fig.set_ylabel(r"${H(X^{(m)})/m$", fontsize=20)       
    fig.set_xlabel(r"$N$", fontsize=20)       
    if do_legend:
        fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    fig.text(1./N_range[1], H0, my_title, fontsize=20)
    
```

```python
## plot results
Fig = plt.figure(figsize=(12,10))

figa= Fig.add_subplot(2,2,1)
#plot_res_vs_N(figa, N_range, res_bins, "Swig bins", 0)
#res_nns_py = np.load("res_nns_py.npy")
plot_res_vs_N(figa, N_range, res_nns_py, "Cython nns", 0)

figa= Fig.add_subplot(2,2,2)
#res_ann_sw = np.load("res_ann_sw.npy")
plot_res_vs_N(figa, N_range, res_ann_sw, "Swig ANN", 1)

figa= Fig.add_subplot(2,2,3)
#res_python = np.load("res_python.npy")
plot_res_vs_N(figa, N_range, res_python, "Python NNS", 0)

figa= Fig.add_subplot(2,2,4)
#res_nns_sw= np.load("res_nns_sw.npy")
plot_res_vs_N(figa, N_range, res_nns_sw, "Swig NNS", 0)
```
